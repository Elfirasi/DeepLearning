### Convertir Bare_Nuclei en num√©rique

```python
#If ‚Äòcoerce‚Äô, then invalid parsing will be set as NaN.
df['Bare_Nuclei'] = pd.to_numeric(df['Bare_Nuclei'],errors='coerce')
```

### <h2 class='text-danger'>Diviser dataset en X (caract√©ristiques) et y (classes)</h2>

```python
X = df.drop(['Class'], axis=1)
y = df['Class']
```

```python
X.head()
```

## Remplacement des val NaN par la m√©diane du colonne

```python
for col in X:
    col_mediane = X[col].median()
    X[col].fillna(col_mediane,inplace=True)
```

```python
X.isnull().sum()
```

### Remplacement des labels par des valeurs binaires

```python
y.replace({2:0,4:1},inplace=True)
```

## D√©couper dataset en 20% pour test et 80% pour traning

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2 )
```

```python
x_train.head()
```

# Normalisation

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)
```

```python
df.info()
```

# R√©seaux de neuronnes

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense 
```

```python
model = Sequential()
model.add(Dense(16,activation='relu', input_dim=9))
model.add(Dense(8,activation='tanh'))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
myloose = tf.keras.losses.BinaryCrossentropy()
optim = tf.keras.optimizers.Adam(learning_rate=0.005)
metrics = ['accuracy']
model.compile(loss=myloose,optimizer=optim,metrics=metrics)
```

```python
history = model.fit(x_train,y_train,validation_split=0.2, epochs=200)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['loss'], label='loss')
plt.grid()
plt.legend()
```

```python
y_pred = model.predict(x_test)
```

```python
import numpy as np
y_pred = np.round(y_pred).astype(int)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score ,f1_score,recall_score , precision_score
cfx=confusion_matrix(y_test,y_pred)
cfx=confusion_matrix(y_test,y_pred)
```
